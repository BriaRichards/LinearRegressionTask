```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Importing the insurance data
ins_data = pd.read_csv("insurance.csv")
```

```python
# Checking the data is loaded correctly
print(ins_data.head())

print(ins_data.describe())
```

```python
# Selecting the rows from the dataframe for plotting a scatterplot 
x = ins_data.age.values
y = ins_data.charges.values

plt.scatter(x, y, color = 'c')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title("Scatterplot to show how age affects insurance costs")
plt.show()
```

```python
# Fitting the model and converting x from 1D array to 2D array
model = LinearRegression()
model.fit(x.reshape(-1,1),y)

# Plotting the model and making predictions
y_pred = model.predict(x.reshape(-1,1))

plt.scatter(x,y,color = 'c')
plt.plot(x,y_pred,color = 'm')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title("Using Linear Regression to show how age affects insurance costs")
plt.show()
```

```python
# Predicting an age to see the effect on insurance cost
unk_x = [[25]] 

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = model.predict(x_pred)

plt.scatter(x,y,color = 'c')
plt.plot(x_pred,y_pred,color = 'm')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title("Using Linear Regression to show how age affects insurance costs")
plt.show()

print("Insurance cost would be around:", model.predict(unk_x))
```
